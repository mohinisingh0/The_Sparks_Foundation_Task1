### Calculating Inter Quartile Range

```python
Q1 = df.quantile(q = 0.25, axis = 0, numeric_only = True, interpolation = 'linear')

Q3 = df.quantile(q = 0.75, axis = 0, numeric_only = True, interpolation = 'linear')
```

```python
IQR = Q3-Q1
IQR
```
